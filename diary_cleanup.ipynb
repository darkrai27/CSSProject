```python
import PyPDF2, nltk, re
from nltk.tokenize import word_tokenize
```

```python
def remove_special_characters(text):
    # Define a regular expression to match Unicode characters
    pattern = re.compile(r'(\\u[0-9A-Fa-f]*|\\n)')

    # Use sub() to replace matched patterns with an empty string
    text = text.replace('\u2009', ' ')
    text = text.replace(' -', '') # Remove end of line hyphenation
    # text = text.replace('\n', 'P_T_S') # Paragraph token separator token to keep paragraph structure
    text = text.replace('\n', ' ')
    # text = text.replace('P_T_S', '\n') # Paragraph token separator token to keep paragraph structure
    cleaned_text = re.sub(pattern, '', text)
    return cleaned_text

def extract_text_pages(pages):
    
    page_texts = [re.sub(r'^\d+\s*', '', remove_special_characters(page.extract_text())) for page in pages]
    return ''.join(page_texts)
```

```python
days = ["MONTAG,", "DIENSTAG,", "MITTWOCH,", "DONNERSTAG,", "FREITAG,", "SAMSTAG,", "SONNTAG,", "VOM"]
pdfFileObj = open('diary.pdf', 'rb')
 
# creating a pdf reader object
pdfReader = PyPDF2.PdfReader(pdfFileObj)
 
# printing number of pages in pdf file
pages = pdfReader.pages[6:-2]

# creating a page object
pageObj = pages[3:7]
text = extract_text_pages(pages)
for day in days:
    text = text.replace(day, '\n' + day)
print(text)
with open('diary-text.txt', 'w') as f:
    f.write(text)
```

```python
lines_array = []

with open('diary-text.txt', 'r') as f:
  for line in f:
    lines_array.append(line)
lines_array
```

```python
text = []
for line in lines_array:
  if "1944 " in line:
    text.append({'text': line.split("1944 ")[-1], "label": "".join(line.split("1944 ")[:-1]).split(",")[0]})
  else: 
    text.append({'text': line.split("1945 ")[-1], "label": "".join(line.split("1945 ")[:-1]).split(",")[0]})
```

```python
for elem in text:
  elem['label'] = ''.join(elem['label']).replace('\n', '')
text[0] = {'text': text[0]['label'], "label": text[0]['text'].replace('\n', '')}
```
